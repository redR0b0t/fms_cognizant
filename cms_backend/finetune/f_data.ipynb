```python
import pandas as pd
data=pd.read_csv('complaints.csv')
# data['complaint'].to_csv('cExamples.csv',index=False)
```

```python

from instruction import instruction
input='complaint: {}'

f_data=pd.DataFrame()
f_data['instruction']=data['complaint'].apply(lambda x: instruction)
f_data['input']=data['complaint'].apply(lambda x:input.format(x))
f_data['output']=data['output']
f_data
```

```python
t_data=f_data.copy()
for i in range(9000):
    t_data=pd.concat([t_data,f_data],axis=0)
t_data
```
